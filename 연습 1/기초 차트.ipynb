```python
#Matplotlib 관련 실습입니다.
#pandas에서 차트를 그릴 때 유용합니다. 

import pandas as pd
import numpy as np
```

```python

import matplotlib.pyplot as plt
#%matplotlib inline #콘솔의 경우
```

```python
plt.figure
plt.plot([1,2,3,4,5,15,7,8,9,10])
plt.show() #주피터노트북, 랩의 경우 실행하지 않아도 표기됩니다. 
```

```python
t = np.arange(0, 10, 0.01)  #0 ~ 12범위까지 0.01 간격(spacing)
y = np.sin(t) #sin값

plt.figure(figsize=(10,3))
plt.plot(t,y)
plt.show()
```

```python
plt.figure(figsize=(10,3))
plt.plot(t, y)
plt.grid() #눈금자 만들기
plt.xlabel('time')
plt.ylabel('Amplitude')
plt.title("Example of sinewave")
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(t, np.sin(t), label='A')
plt.plot(t, np.cos(t), label='B')
plt.grid()
plt.xlabel('aaaaa')
plt.ylabel('bbbbb')
plt.title("only english / no korean")
plt.legend()
plt.show()
```
